```python
def get_user_id(conn):
    cur = conn.cursor()
    sql = 'SELECT payID FROM trade_guangzhou GROUP BY payID HAVING count(*) > 15;'
    cur.execute(sql)
    ids = cur.fetchall()
    cur.close()
    return ids
```

```python
def get_user_info(conn, the_id):
    cur = conn.cursor()
    cur.execute('select * from  trade_guangzhou where payID = %s', (the_id))
    info = cur.fetchall()
    cur.close()
    info_list = [{
        'payID' : item[9],
        'stationStart' : item[5],
        'stationEnd' : item[6],
        'tradeTime' : pd.Timestamp(item[7]), 
        } for item in info]
    return pd.DataFrame(info_list)
```

```python
def norm(counter):
    the_dict = dict(counter)
    if '\N' in the_dict:
        del the_dict['\N']
#     if '虫雷港' in the_dict:
#         del the_dict['虫雷港']
    sum_values = sum(the_dict.values())
    for key, value in the_dict.items():
        the_dict[key] = float(value) / sum_values
    return the_dict

def analysis_user(records):
    counter_station_week = Counter()
    counter_station_weekend = Counter()
    counter_station_may_home = Counter()
    
    week_time = []
    weekend_time = []
        
    for i in range(records.shape[0]):
        start_station = records.loc[i, 'stationStart']
        end_station = records.loc[i, 'stationEnd']
        trade_time = records.loc[i, 'tradeTime']
        
        time = None
        time_record = route[route['From'] == start_station][route['To'] == end_station]['Time']
        if not time_record.empty:
            time = time_record.values[0]
        if trade_time.isoweekday() in [6, 7]:
            counter_station_weekend.update([start_station, end_station])
            if time:
                week_time.append(time)
        else:
            counter_station_week.update([start_station, end_station])
            if time:
                weekend_time.append(time)
        if trade_time.hour < 10:
            counter_station_may_home.update(start_station)
        if trade_time.hour > 17:
            counter_station_may_home.update(end_station)
    
    pro_station_week =  norm(counter_station_week)
    pro_station_weekend = norm(counter_station_weekend)
    pro_station_may_home = norm(counter_station_may_home)
    return pro_station_week, pro_station_weekend, pro_station_may_home, week_time, weekend_time

def get_all_users_matrix(conn, ids, station_names):
    the_ids = [i[0] for i in ids]
    
    station_week = pd.DataFrame(np.zeros([len(ids), len(station_names)]),columns=station_names, index=np.array(the_ids))
    station_weekend = pd.DataFrame(np.zeros([len(ids), len(station_names)]),columns=station_names,index=np.array(the_ids))
    station_may_home= pd.DataFrame(np.zeros([len(ids), len(station_names)]),columns=station_names,index=np.array(the_ids))
    time_week = pd.DataFrame(np.zeros([len(ids), 2]), columns=['time', 'average'], index=np.array(the_ids))
    time_weekend = pd.DataFrame(np.zeros([len(ids), 2]), columns=['time', 'average'], index=np.array(the_ids))
    
    for the_id in ids:
        info = get_user_info(conn, the_id)
        pro_station_week, pro_station_weekend, pro_station_may_home, week_time, weekend_time = analysis_user(info)
        
        if week_time:
            time_week.loc[the_id[0], 'time'] = ' '.join([str(i) for i in week_time])
            time_week.loc[the_id[0], 'average'] = np.array(week_time).mean()
        if weekend_time:
            time_weekend.loc[the_id[0], 'time'] = ' '.join([str(i) for i in weekend_time])    
            time_weekend.loc[the_id[0], 'average'] = np.array(week_time).mean()
        
        for key, value in pro_station_week.items():
            if key in station_names:
                station_week.loc[the_id[0], key] = value
        for key, value in pro_station_weekend.items():
            if key in station_names:
                station_weekend.loc[the_id[0], key] = value
        for key, value in pro_station_may_home.items():
            if key in station_names:
                station_may_home.loc[the_id[0], key] = value
        
    return station_week, station_weekend, station_may_home, time_week, time_weekend
```

```python
route = pd.read_csv('../route.csv')
ids = get_user_id(conn)
station = pd.read_csv('../station_env.csv')
station_names = [i.decode('utf-8') for i in station['name']]
```

```python
station_week_matrix, station_weekend_matrix, station_may_home_matrix, time_week, time_weekend = get_all_users_matrix(conn, ids, station_names)
```
