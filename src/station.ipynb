```python
data = pd.DataFrame(data = station['name'])
```

```python
def add_feature(data, feature_name, feature_data):
    data[feature_name] = np.zeros(data.shape[0])
    for index, row in data.iterrows():
        records = feature_data[feature_data['subway'] == row['name']]
        val = 0.0
        for i, record in records.iterrows():
            val += float(record['overall_rating']) / ((record['distance'] + 1) / 100.)
        if val >= 0:
            data.loc[index, feature_name] = val
    return data
```

```python
data= add_feature(data, 'life', life)
data= add_feature(data, 'hotel', hotel)
data= add_feature(data, 'place', place)
data= add_feature(data, 'cater', cater)
```
